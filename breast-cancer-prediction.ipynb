# Breast Cancer Prediction Using Machine Learning

# Importing libraries


```python
import pandas as pd
import seaborn as sns
import numpy as np
```

# Loading and Displaying data


```python
df = pd.read_csv("data.csv")

df.tail()
```

```python
# counting rows and colums
df.shape
```

```python
# count number of null (empty values) 
df.isna().sum()
```

```python
# drop the column with null values
df.dropna(axis=1, inplace=True)
```

```python
# new count of rows and colums
df.shape
```

```python
# get count of number of M or B cells in diagnosis
df['diagnosis'].value_counts()
```

```python
# get datatypes of each column
df.dtypes
```

```python
# encoding diagnosis values
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df.iloc[:,1]  = labelencoder.fit_transform(df.iloc[:,1].values) # substitutes M and B for 1 and 0
```

```python
# displaing df
df
```

### Spliting dataset & Feature Scaling


```python
# splitting dataset into independent and dependent datasets
x = df.iloc[:,2:].values
y = df.iloc[:,1].values
y = y.astype('int')
```

```python
# splitting datasets into training (75%) and testing (25%)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
x_train
```

# Making predictions with Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
m1 = DecisionTreeClassifier()
```

```python
m1.fit(x_train, y_train)
```

```python
m1.score(x_test, y_test)
```

# Making predictions with SVM

```python
from sklearn.svm import SVC
m2 = SVC()
m2.fit(x_train,y_train)
m2.score(x_test, y_test)
```

## Improving the score using K-Cross-Fold

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(SVC(), x_train, y_train, cv = 5)
scores.max()
```
